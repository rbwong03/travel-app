```python
from utils import cities, first_names, last_names, adjectives
import random
import hashlib
import string
from datetime import datetime, timedelta


def get_random_date(start_year=2000, end_year=2024):
    random_datetime = get_random_datetime(start_year, end_year)
    return random_datetime.date()


def get_random_datetime(start_year=2000, end_year=2024):
    start_date = datetime(start_year, 1, 1, 0, 0, 0)
    end_date = datetime(end_year, 12, 31, 23, 59, 59)
    
    time_diff = end_date - start_date
    
    random_seconds = random.randint(0, int(time_diff.total_seconds()))
    random_date = start_date + timedelta(seconds=random_seconds)
    
    return random_date


def get_datetime_within_next_day(given_date_str):
    
    given_date = datetime.strptime(given_date_str, '%Y-%m-%d')
    
    start_datetime = given_date + timedelta(days=1)
    
    random_datetime = start_datetime + timedelta(seconds=random.randint(0, 86399))
    
    return random_datetime

def random_decimal(min_val=0, max_val=10, precision=1):
    range_start = min_val * 10**precision
    range_end = max_val * 10**precision
    

    random_num = random.randint(range_start, range_end) / 10**precision
    
    return random_num

def make_salt():
    char_set = string.printable[:95].replace("'", "").replace('"', "")
    salt = ''.join(random.choice(char_set) for _ in range(8))
    return salt


def hash_password_with_salt(salt):
    salted_password = 'password' + salt
    hash_obj = hashlib.sha256()
    
    hash_obj.update(salted_password.encode('utf-8'))
    
    return hash_obj.hexdigest()

```

```python
import random

def generate_connections_inserts(num_users, max_connections_per_user):
    insert_statements = []
    for user_id1 in range(1, num_users + 1):
        num_connections = random.randint(1, max_connections_per_user)
        
        connections = random.sample(range(1, num_users + 1), num_connections)
        connections = [user_id2 for user_id2 in connections if user_id2 != user_id1]
        
        for user_id2 in connections:
      
            smaller_id, larger_id = sorted([user_id1, user_id2])
            insert_statement = f"INSERT INTO connections (user_id1, user_id2) VALUES ({smaller_id}, {larger_id});"
            
            if insert_statement not in insert_statements: 
                insert_statements.append(insert_statement)

    return insert_statements

num_users = 20 
max_connections_per_user = 15 
insert_statements = generate_connections_inserts(num_users, max_connections_per_user)

for statement in insert_statements:
    print(statement)
```
